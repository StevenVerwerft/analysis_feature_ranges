 ## Plan van aanpak
 
De analyse kan zowel globaal over alle companies lopen als per specifieke instances. In de globale analyse gaan we de completeness per feature na, en zullen we deze completeness analyseren over de tijd.

Nuttige metrieken om een globale analyse te doen:

Per meetpunt in de tijd (wekelijks)

* Minimale en maximale waarde van de feature
* Gemiddelde waarde en standaardafwijking van de feature
* Proportie missing values van de feature

```python
import os


files = [f for f in os.listdir(path='data_teamleader/') if f.endswith('.csv')]
print(files)
dates = [name.strip('.csv').strip('teamleader_') for name in files]
print(dates)
dataframes = {date : pd.read_csv('data_teamleader/'+file, sep=';', thousands=',', index_col=1) for date, file in zip(dates, files)}
#s = pd.Series(dates)
#s = pd.to_datetime(s, format='%d%m%y')

feature_names = ['name', 'employees (upper bound)','revenue', 'datascouts score', 'social_score',
       'traction_score', 'business_size_score', 'seo_score',
       'median_load_time', 'median_load_time_percentile',
       'incoming_links_count', 'alexa_rank', 'alexa_rank_change',
       'reach_alexa_rank_change', 'reach_per_million',
       'reach_per_million_change_percentile', 'page_views_per_user',
       'page_views_per_user_change_percentile', 'page_views_per_million',
       'page_views_per_million_change_percentile', 'page_views_alexa_rank',
       'page_views_alexa_rank_change', 'rank', 'top_rank',
       'domain_rating', 'backlinks_count', 'referring_domains_count',
       'referring_ips_count', 'referring_class_c_count',
       'referring_pages_count', 'checkins_count', 'likes_count',
       'talking_about_count', 'listed_count', 'tweet_count',
       'favourites_count']

dataframes = {date : dataframes[date][feature_names] for date in dates}
#dataframes = {date : dataframes[date]['date'] for date in dates}

for key in dataframes.keys():
    dataframes[key]['date'] = pd.to_datetime(key, format='%d%m%y')
    dataframes[key] = dataframes[key].rename(index=str, columns={'employees (upper bound)' : 'employees'})




# om filesize te meten
import sys
import _pickle
mystring = _pickle.dumps(dataframes)

print(round(sys.getsizeof(mystring) / 1e6, 2), 'MB')
```

```python
# Employee size

print(dataframes['010817'].columns)

global_features = ['employees', 'revenue', 'datascouts score', 'social_score',
       'traction_score', 'business_size_score', 'seo_score',
       'median_load_time', 'median_load_time_percentile',
       'incoming_links_count', 'alexa_rank', 'alexa_rank_change',
       'reach_alexa_rank_change', 'reach_per_million',
       'reach_per_million_change_percentile', 'page_views_per_user',
       'page_views_per_user_change_percentile', 'page_views_per_million',
       'page_views_per_million_change_percentile', 'page_views_alexa_rank',
       'page_views_alexa_rank_change', 'rank', 'top_rank', 'domain_rating',
       'backlinks_count', 'referring_domains_count', 'referring_ips_count',
       'referring_class_c_count', 'referring_pages_count', 'checkins_count',
       'likes_count', 'talking_about_count', 'listed_count', 'tweet_count',
       'favourites_count']


pivot_tables = {}
for i in range(len(global_features)):
    print(global_features[i])
    features = ['name', 'date', global_features[i]]
    df = dataframes['010817'][features]
    for key in list(dataframes.keys())[1: ]:
        df = df.append(dataframes[key][features])

    pivot_tables.update({global_features[i] : pd.pivot_table(df, values=global_features[i], index='date', columns='name')})

```

```python
import pickle
pickle.dump(pivot_tables, open('pivot_tables.p', 'wb'))
```
